```python
startup=pd.read_csv("50_Startups.csv")
```

```python
df=startup.rename({'R&D Spend':'RDS','Administration':'Admin','Marketing Spend':'MS'},axis=1)
```

```python
X=df.iloc[:,0:3]
y=df.iloc[:,-1]
```

```python
# Importing train_test_split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=0)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X.describe()
```

```python
array1=X_train.values
```

```python
# Scaling our data

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train[['RDS','Admin','MS']]=scaler.fit_transform(array1)
```

```python
X_train
```

```python
array2=X_test.values
```

```python
X_test[['RDS','Admin','MS']]=scaler.fit_transform(array2)
X_test
```

```python
print("Scaled Training DataFrame\n",X_train,"\n\nScaled Testing DataFrame\n",X_test) 
```

```python
from sklearn.linear_model import LinearRegression
regressor_best=LinearRegression()
regressor_best.fit(X_train,y_train)
```

```python
regressor_best
```

```python
regressor_best.intercept_
```

```python
regressor_best.coef_
```

```python
y_pred_test=regressor_best.predict(X_test)
y_pred_test
```

```python
print(y_test)
```

```python
y_pred_train=regressor_best.predict(X_train)
y_pred_train
```

```python
print(y_train)
```

```python
import numpy as np
```

```python
from sklearn.metrics import r2_score

R2_train=r2_score(y_train,y_pred_train)
R2_test=r2_score(y_test,y_pred_test)

print("R-squared Train",np.round(R2_train,5),"\nR-squared Test",np.round(R2_test,5))
```

```python
rsq=pd.DataFrame({'Test':['Training','Testing'],'R-Squared':[np.round(R2_train,5),np.round(R2_test,5)]})
rsq
```
